```python
import requests # 获取所有网页内容
from bs4 import BeautifulSoup # 解析标签内容

# 获取网页内容必备的headers和cookies
# 右键检查，使用网络任一链接获取
headers = {
    'User-Agent': 'xxx'
}

cookies = {'cookie':'xxx'}

url = 'https://5sing.kugou.com/fc/16759037.html'  # 请将此处替换为实际的网页URL
response = requests.get(url,headers=headers,cookies=cookies)
web_content = response.text

print(web_content)
```

```python
soup = BeautifulSoup(web_content, 'html.parser')
target_content = soup.find('div', {'class' : 'lrc_info_clip inspiration-tab-content'})
lyrics = target_content.text
print(lyrics)

with open('output_data/basic_demo.txt', 'w', encoding = 'gb18030') as file:
	file.write(lyrics)
```
